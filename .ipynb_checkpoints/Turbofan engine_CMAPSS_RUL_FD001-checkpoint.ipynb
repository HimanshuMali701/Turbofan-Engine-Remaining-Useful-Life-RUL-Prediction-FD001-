# Turbofan Engine Remaining Useful Life (RUL) Prediction — FD001

This notebook demonstrates step-by-step development of a machine-learning‐based Remaining Useful Life (RUL) prediction model using the **NASA CMAPSS FD001 dataset**.

---


## Import Libraries and Set Configuration
This cell imports the required Python libraries and sets the working directory.  
It must be executed **before running any further cells**.


```python
import os
import numpy as np
import pandas as pd

import matplotlib.pyplot as plt
import seaborn as sns

from sklearn.preprocessing import MinMaxScaler
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score

import joblib  

sns.set(style="whitegrid")
plt.rcParams["figure.figsize"] = (10, 6)

DATA_DIR = "data"   
DATASET_ID = "FD001"  # main dataset for now
RANDOM_STATE = 42
```

**Interpretation**

 This cell loads all required packages for data handling, visualization, and machine learning.


## Load Training Dataset

We define a helper function to correctly load CMAPSS text files and assign column names.


```python
def load_cmapss_dataset(dataset_id="FD001", split="train"):
    """
    Load CMAPSS dataset (FD00x) from text file.
    split: 'train' or 'test'
    """
    assert split in ["train", "test"], "split must be 'train' or 'test'"
    filename = f"{split}_{dataset_id}.txt"
    path = os.path.join(DATA_DIR, filename)

    col_names = (
        ["unit", "cycle", "os1", "os2", "os3"] +
        [f"s{i}" for i in range(1, 22)]  
    )

    df = pd.read_csv(
        path,
        sep=r"\s+",
        header=None,
        names=col_names
    )

    print(f"{split}_{dataset_id} shape:", df.shape)
    print(df.head())
    return df

train_df_raw = load_cmapss_dataset(dataset_id=DATASET_ID, split="train")
```

**Interpretation**

- The dataset contains multiple engines (`unit`) with time-series cycles.
- The first 3 columns are operational settings, and sensor readings follow (`s1–s21`).
- `.head()` confirms correct loading and column labeling.




## Compute Remaining Useful Life (RUL)

For each engine, RUL = *(last cycle) − (current cycle)*.


```python
def add_rul(df):
    max_cycles = df.groupby("unit")["cycle"].max().reset_index()
    max_cycles.columns = ["unit","max_cycle"]

    df = df.merge(max_cycles, on="unit", how="left")
    df["RUL"] = df["max_cycle"] - df["cycle"]
    df.drop(columns=["max_cycle"], inplace=True)
    return df

train_df = add_rul(train_df_raw)
train_df.head()
```

**Interpretation**

- RUL decreases from a maximum value to 0 as the engine approaches failure.
- Each row now includes a target value (`RUL`) used for model training.




## Explore Dataset & Sensor Behavior

Understanding data distributions helps identify useful features and remove noise.


```python
sensor_cols = [c for c in train_df.columns if c.startswith('s')]

desc = train_df[sensor_cols].describe().T
desc["nunique"] = train_df[sensor_cols].nunique()

plt.figure()
sns.histplot(train_df["RUL"], bins=50, kde=True)
plt.title("RUL Distribution")
plt.show()

desc.head(10)
```

**Interpretation**

- The histogram shows how failure cycles vary across engines.
- Columns with extremely low variance (`std ≈ 0`) or low unique values carry no meaningful patterns and will be removed.



## Drop Low-Variance Sensors


```python
sensors_to_drop = ['s1', 's5', 's10', 's16', 's18', 's19']
train_df_reduced = train_df.drop(columns=sensors_to_drop)

print("Dropped Sensors:", sensors_to_drop)
train_df_reduced.head()
```

**Interpretation**

- These sensors showed nearly constant values and do not contribute to prediction.
- Removing them reduces noise and improves model performance.




## Normalize Data (Min-Max Scaling)

Scaling ensures all sensors contribute equally during training.


```python
feature_cols = ['os1','os2','os3'] + [c for c in train_df_reduced.columns if c.startswith('s')]
target = "RUL"

scaler = MinMaxScaler()
X_train_scaled = scaler.fit_transform(train_df_reduced[feature_cols])

train_scaled = pd.DataFrame(X_train_scaled, columns=feature_cols)
train_scaled["RUL"] = train_df_reduced["RUL"]

train_scaled.head()
```

**Interpretation**

- All feature values now lie between **0 and 1**, allowing the machine learning algorithm to learn efficiently.
- This scaled dataset will be used to train the predictive model.



## Split Data into Training & Validation Sets

To evaluate the model before final testing, we create a validation set from the processed training data.


```python
from sklearn.model_selection import train_test_split

X = train_scaled[feature_cols].values
y = train_scaled["RUL"].values

X_train, X_val, y_train, y_val = train_test_split(
    X, y, test_size=0.2, random_state=RANDOM_STATE
)

print("Training shape:", X_train.shape)
print("Validation shape:", X_val.shape)
```

**Interpretation**

- The dataset is split into:
  - **80% for training**
  - **20% for validation**
- The validation set helps measure generalization ability **before testing on unseen engines**.




##  Train Machine Learning Model (Random Forest)

RandomForestRegressor is chosen because it works well on tabular sensor data and allows feature importance analysis.


```python
rf_model = RandomForestRegressor(
    n_estimators=200,
    random_state=RANDOM_STATE,
    n_jobs=-1
)

rf_model.fit(X_train, y_train)

print("Model training complete.")
```

**Interpretation**

- A **Random Forest** with 200 decision trees is trained.
- The model has now learned to map sensor behavior to expected remaining engine lifespan.



## Validate Model Performance

We now check how well the model performs on unseen validation samples.


```python
val_pred = rf_model.predict(X_val)

val_mae = mean_absolute_error(y_val, val_pred)
val_rmse  = mean_squared_error(y_val, val_pred) ** 0.5
val_r2 = r2_score(y_val, val_pred)

print(f"Validation MAE   : {val_mae:.3f}")
print(f"Validation RMSE  : {val_rmse:.3f}")
print(f"Validation R²    : {val_r2:.3f}")
```

**Interpretation**

- **MAE (Mean Absolute Error):** How far predictions are from ground truth (lower = better)
- **RMSE:** Penalizes large mistakes (lower = better)
- **R² Score:** Measures how well model explains variance (1.0 = perfect)



## Load Test Data and True RUL Values

Now we repeat preprocessing steps for the test data so it matches the training format.


```python
test_df_raw = load_cmapss_dataset(DATASET_ID, "test")

# load true RUL for last cycle per engine
rul_path = os.path.join(DATA_DIR, f"RUL_{DATASET_ID}.txt")
rul_df = pd.read_csv(rul_path, header=None, names=["RUL_last"])

rul_df.head()
```

**Interpretation**

- The test dataset includes engines at unknown degradation levels.
- The separate file `RUL_FD001.txt` provides only the final ground truth value — we must reconstruct full cycle-level RUL next.


##  Compute Remaining Useful Life for Test Cycles


```python
def add_test_rul(df, rul_df):
    rul_df["unit"] = np.arange(1, rul_df.shape[0]+1)

    max_cycles = df.groupby("unit")["cycle"].max().reset_index()
    max_cycles.columns = ["unit", "max_cycle"]

    df = df.merge(max_cycles, on="unit", how="left").merge(rul_df, on="unit", how="left")

    df["RUL"] = (df["max_cycle"] - df["cycle"]) + df["RUL_last"]
    df.drop(columns=["max_cycle", "RUL_last"], inplace=True)

    return df

test_df = add_test_rul(test_df_raw, rul_df)
test_df.head()
```

**Interpretation**

- This step reconstructs full RUL progression for each test engine.
- The test set is now properly labeled and ready for prediction evaluation.



## Match Test Data to Training Format



```python
test_df_reduced = test_df.drop(columns=sensors_to_drop)

X_test_scaled = scaler.transform(test_df_reduced[feature_cols])

y_test = test_df_reduced["RUL"].values
```

**Interpretation**

- Consistency is crucial: train and test data must contain identical columns formatted the same way.
- We reuse the trained scaler — we **never fit again** to avoid data leakage.




## Final Model Evaluation on Test Set



```python
test_pred = rf_model.predict(X_test_scaled)

test_mae = mean_absolute_error(y_test, test_pred)
test_rmse  = mean_squared_error(y_test, test_pred) ** 0.5
test_r2 = r2_score(y_test, test_pred)

print(f"TEST MAE  : {test_mae:.3f}")
print(f"TEST RMSE : {test_rmse:.3f}")
print(f"TEST R²   : {test_r2:.3f}")
```

**Interpretation**

These are your **official final project results**.




## Feature Importance Analysis

This identifies which sensors influence RUL prediction.


```python
importance = pd.Series(rf_model.feature_importances_, index=feature_cols).sort_values(ascending=False)

plt.figure(figsize=(8,6))
sns.barplot(x=importance.values[:15], y=importance.index[:15])
plt.title("Top Most Important Sensors")
plt.show()
```

**Interpretation**

- The most influential sensors may indicate early degradation patterns.
- This insight helps explain model behavior and supports engineering interpretation.



## Save Model for Deployment 


```python
joblib.dump(rf_model, "rf_model_FD001.joblib")
joblib.dump(scaler, "scaler_FD001.joblib")

print("Model and scaler saved successfully.")
```
